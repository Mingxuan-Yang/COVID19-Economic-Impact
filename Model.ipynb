```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.tree import DecisionTreeRegressor
```

```python
merged = pd.read_csv('merged.csv')
merged.head()
```

```python
pca = PCA()
X = pca.fit_transform(merged[['unemp rate', 'Q2 2020']])
merged['PCA'] = X[:,0]
```

```python
pca.explained_variance_ratio_
```

```python
cities = merged.sort_values('PCA')[:5]
cities
```

```python
merged.sort_values('PCA', ascending = False)[:5]
```

```python
clf = DecisionTreeRegressor()
features = ['industry', 'Density', 'Total Cases', 'Total Deaths']
clf.fit(merged[features], merged['PCA'])
for i in range(len(features)):
    print(f'Feature Importance of {features[i]}', round(clf.feature_importances_[i], 4))
```
